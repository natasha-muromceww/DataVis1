```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(
    FigureClass=Waffle,
    rows=12,
    columns=12, 
    values={'tops':57, 'bottoms':33, 'accessories':30, 'outerwear':16, 'shoes':12},
    colors=["#e0218a", "#ff3399", "#ff70b8", "#ffadd6", "#ffd6eb"],
    icons=['star','heart', 'sun', 'circle', 'square'],
    font_size=15,
    icon_legend=True,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)})

plt.title('The amount of each type of item in my closet per my latest count', fontsize=25)


                         
```
