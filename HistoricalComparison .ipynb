# Historical Comparison of Natasha's Sleep per Month

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['Week 1', 'Week 2', 'Week 3', 'Week 4']
september_sleep = [7.13, 7.20, 7.16, 7.36]
october_sleep = [6.93, 6.31, 7.11, 6.06]
november_sleep = [6.08, 6.63, 6.45, 9.15]
```

```python
x = np.arange(len(labels)) 
width = 0.2  

fig, ax = plt.subplots()
rects1 = ax.bar(x - 2 * (width/3), september_sleep, width, label='September')
rects2 = ax.bar(x + (width/3), october_sleep, width, label='October')
rects3 = ax.bar(x + 4 * (width/3), november_sleep, width, label='November')

ax.set_ylabel('Sleep (hours)')
ax.set_xlabel('Average Sleep per Week of the Month')
ax.set_title('Sleep by Month and Week')
ax.legend()

fig.tight_layout()
plt.show()
```
