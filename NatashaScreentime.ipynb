# DataVis Natasha's Screentime

Using Pie charts to look at my screentime between apps on my iphone. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Week 1 (11/28 - 12/4)

```python
# Writing out time spent per app
hours = 60 
#Week_1: 11/28 - 12/4
week_1_total = 37 * hours + 32
week_1_Snapchat = 8 * hours + 32
week_1_Spotify = 6 * hours + 43
week_1_TikTok = 5 * hours + 13
week_1_Instagram = 3 * hours + 24
week_1_Messages = 2 * hours + 47
week_1_Twitter = 1 * hours + 18
week_1_Other = week_1_total - (week_1_Snapchat + week_1_Spotify 
                               + week_1_TikTok + week_1_Instagram 
                               + week_1_Messages + week_1_Twitter)
```

```python
#plotting pie chart of time spend on each app 
df1 = pd.DataFrame([week_1_Snapchat, week_1_Spotify, week_1_TikTok, week_1_Instagram, 
                   week_1_Messages, week_1_Twitter, week_1_Other], 
                  index=['Snapchat', 'Spotify', 'TikTok', 'Instagram', 'Messages', 'Twitter', 'Other'], 
                  columns=['Apps'])
 
df1.plot(kind='pie', subplots=True, figsize=(6, 6))
plt.show()
```
